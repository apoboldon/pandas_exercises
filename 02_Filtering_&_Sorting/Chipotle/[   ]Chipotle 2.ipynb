### Step 4. How many products cost more than $10.00?

### [   ] ¿está bien???????

```python
df.item_price=df.item_price.replace('\$', '', regex=True)
```

```python
df.item_price.dtype
```

```python
df.item_price= df.item_price.astype(float)
```

```python
df.item_price.dtype
```

```python
df[df.item_price>10]
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

### [   ] ¿cómo la haría?????????????

```python
dfsel=df[df.quantity==1]
dfsel=dfsel.drop_duplicates(subset= 'choice_description',ignore_index=True)
dfsel
```

```python
dfsel.sort_values(by='item_price',ascending=True)
```

### Step 6. Sort by the name of the item

```python
dfsel.sort_values(by='item_name').head(20)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
dfsel=df[['item_name','item_price']]
dfsel.sort_values(by='item_price', ascending=False).head(1)
```

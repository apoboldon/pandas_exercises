### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users=pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep= '|', index_col= 'user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by='occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
dfsel=pd.get_dummies(users, columns=['gender'])
dfsel.groupby(by='occupation').gender_M.mean() # con pd.get_dummies asignamos 0 o 1 a si es hombre o mujer.

percent=(dfsel.groupby(by='occupation').gender_M.mean())*100 # si queremos ver los resultados en % lo multiplicamos por 100.
percent
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(by='occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
dfsel=pd.get_dummies(users, columns=['gender'])
M=dfsel.groupby(by='occupation').gender_M.mean()
M
F=dfsel.groupby(by='occupation').gender_F.mean()
F

Male_Female=pd.concat([F,M],axis=1).reset_index() # con la funciÃ³n .concat, ponemos las dos columnas que queremos observar.
Male_Female
```

```python
Male_Female[(Male_Female['gender_F']*100), (Male_Female['gender_M']*100)] # asÃ­ lo harÃ­amos para que salier en % base 100, pero como hay un dato que es 0, al multiplicarlo por 100 da error
```

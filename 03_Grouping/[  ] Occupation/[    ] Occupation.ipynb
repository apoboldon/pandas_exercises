### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
df=pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep= '|', index_col= 'user_id')
df
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby('occupation').age.mean().reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

### [   ] Manera mÃ¡s sencilla de hacerlo????????????

```python
dfM=df[df.gender=='M']
dfM=dfM.groupby('occupation').gender.count().reset_index()
dfM.rename(columns={'gender': 'gender M'}, inplace= True )
dfM.head()
```

```python
dfF=df[df.gender=='F']
dfF=dfF.groupby('occupation').gender.count().reset_index()
dfF.rename(columns={'gender': 'gender F'}, inplace= True )
dfF.head()
```

```python
dfMF=pd.concat([dfM,dfF], axis=1)
dfMF=dfMF.iloc[:, [0,1,3]]
dfMF.head()
```

```python
dfMF['percent M']= dfMF['gender M']/ (dfMF['gender F'] + dfMF['gender M'])*100
round(dfMF).head(10)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

### [  ] Â¿se pude hacer de otra manera???????

```python
dfsel=df.groupby(['occupation', 'gender']).age.mean().reset_index()
dfsel=pd.pivot(dfsel,index= 'occupation', columns= 'gender', values= 'age')
round(dfsel)
```
